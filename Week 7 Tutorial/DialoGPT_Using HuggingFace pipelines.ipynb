# DialoGPT- CHATBOT
Let's create a simple chatbot using Hugging Face's Dialogpt model

*   Documentation: https://huggingface.co/transformers/model_doc/dialogpt.html
*   Model details: https://arxiv.org/pdf/1911.00536.pdf



```
!pip install transformers==4.22.2

from transformers import AutoModelForCausalLM, AutoTokenizer  #Loading the pre-trained DialoGPT model for causal language modeling and the Tokenizer associated with the DialoGPT model
import torch #Importing the PyTorch library
```

```
# Load pre-trained model and tokenizer for the DialoGPT-medium model
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")
```

```
# Initialize chat history to none. This variable will store the conversation history between the user and the bot
chat_history_ids = None

# Chat loop for 5 iterations
for step in range(5):
    # Get user input
    user_input = input("Me 😎: ")

    # Encode user input into token IDs and add end of string token. The return_tensors='pt' argument ensures that the output is a PyTorch tensor
    new_user_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

    # Append user input to chat history to maintain context across the conversation. Without it, the bot will not hold prior memory of the conversations
    # Used an if statement to check if chat_history_ids is None. If it is, set bot_input_ids to new_user_input_ids. Otherwise, concatenate chat_history_ids with new_user_input_id
    if chat_history_ids is None:
        bot_input_ids = new_user_input_ids
    else:
        bot_input_ids = torch.cat([chat_history_ids, new_user_input_ids], dim=-1) #concatenate the new user input with the existing chat history

    # Generate response
    chat_history_ids = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

    # Decode and print response
    bot_response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)
    print("Bot 🤖: {}".format(bot_response))
```
